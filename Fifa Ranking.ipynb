```python
from requests import get
from bs4 import BeautifulSoup
```

```python
url = 'https://www.fifa.com/fifa-world-ranking/ranking-table/men/'
```

```python
response = get(url)
response
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
html_soup
```

```python
top_heading=html_soup.find_all('div',class_='fi-rankings-disclaimer');
heading_data=top_heading[0].text
heading_data=heading_data.replace("\r\n"," ")
heading_data
```

```python
table_heading=html_soup.find_all('thead')
table_heading_data=html_soup.find_all('span',class_="th-text")
ranking_table_data=html_soup.find_all('tbody')
```

```python
for x in table_heading_data:
    print(x.text.replace("\r\n",""),end="")
for i,r in enumerate(ranking_table_data):
    name_list=r.find_all('span',class_="fi-t__nText")
    points_list=r.find_all('td',class_="fi-table__td fi-table__points")
    prev_point_list=r.find_all('td',class_="fi-table__td fi-table__prevpoints")
    rank_mov_list=r.find_all('td',class_="fi-table__td fi-table__rankingmovement")   
```

```python
print("\t\t",heading_data)
print("\n\n")
for x in table_heading_data:
    print(x.text.replace("\r\n"," "),end="")
print("\n")
for i, (name,points,prev,rankmov) in enumerate(zip(name_list, points_list,prev_point_list,rank_mov_list)):
    print('{0:>6}{1:^25}{2:^10}{3:^30}{4:^5}'.format(i+1,name.text,points.text,prev.text,rankmov.text))
    
    
    
```
